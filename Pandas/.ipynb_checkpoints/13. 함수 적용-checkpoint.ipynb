# 13. 함수 적용


```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df['학교'] = df['학교'] + '등학교'
df
```

```python
df['키'] = df['키'] + 'cm'
df
```

## 데이터에 함수 적용 (apply)

```python
# 키 뒤에 cm 을 붙이는 역할
def add_cm(height):
    return str(height) + 'cm'

df['키'] = df['키'].apply(add_cm)
df
```

```python
def capitalize(lang):
    if pd.notnull(lang):
        return lang.capitalize()
    return lang

df['SW특기'] = df['SW특기'].apply(capitalize)
df
```
